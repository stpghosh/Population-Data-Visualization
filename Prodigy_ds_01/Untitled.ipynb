```python
# This part assumes you've previously merged male and female data
# and then selected the top 10 countries based on total population.

# Sample merged_data (similar to what you'd have after the merge step)
merged_data_sample = pd.DataFrame({
    'Country Code': ['CHN', 'IND', 'USA', 'IDN', 'PAK', 'NGA', 'BRA', 'BGD', 'RUS', 'MEX', 'JPN', 'ETH'],
    '2022_male':    [710000000, 700000000, 160000000, 137000000, 115000000, 109000000, 107000000, 84000000, 70000000, 63000000, 61000000, 60000000],
    '2022_female':  [700000000, 680000000, 170000000, 138000000, 116000000, 109000000, 108000000, 84000000, 75000000, 64000000, 64000000, 60000000],
    'Total population': [1410000000, 1380000000, 330000000, 275000000, 231000000, 218000000, 215000000, 168000000, 145000000, 127000000, 125000000, 120000000]
})

# Assuming 'sorted_data' was created by sorting 'merged_data_sample' by 'Total population'
sorted_data = merged_data_sample.sort_values(by="Total population", ascending=False)

# Get the top 10 countries based on the 'Total population' (important for 'top_10_countries')
top_10_countries = sorted_data.head(10)

print("Top 10 Countries Data (for plotting):")
print(top_10_countries[['Country Code', '2022_male', '2022_female', 'Total population']])
print("-" * 50)

# --- 2. Create the Stacked Bar Plot (Your original code, slightly refined) ---

plt.figure(figsize=(12, 6))

# Plot female population (bottom layer)
sns.barplot(
    x="Country Code",
    y="2022_female",
    data=top_10_countries,
    color="red",
    label="Female Population"
)

# Plot male population on top of female population
sns.barplot(
    x="Country Code",
    y="2022_male",
    data=top_10_countries,
    # 'bottom' parameter is crucial for stacking, it starts the 'male' bar where 'female' bar ends
    bottom=top_10_countries["2022_female"],
    color="green",
    label="Male Population"
)

plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.title("Male vs. Female Population in Top 10 Countries (2022)", fontsize=14)
plt.legend(title="Gender") # Added a title to the legend
plt.xticks(rotation=45, ha="right", fontsize=10) # Aligned tick labels for readability
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7) # Add a subtle grid for better readability

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
